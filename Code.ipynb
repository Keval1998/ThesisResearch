```python
import numpy as np
import skvideo.io
import glob
import cv2
def create_video(dir_path, name):
  paths = glob.glob(dir_path+"/*.png")
  paths_sorted = sorted(paths, key = lambda x: int(x.split("/")[-1].split(".")[0]))
  print(paths_sorted)
  writer = skvideo.io.FFmpegWriter(name)
  for i in range(len(paths)):
      img = cv2.imread(paths_sorted[i])
      img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
      writer.writeFrame(img)
  writer.close()
```

Our UNET like architecture, uncomment and run two cells after u_net definitions to study network architecture in details

```python
def u_net(input_shape, dropout_rate, l2_lambda):
  
  # Encoder
  input = Input(shape = input_shape, name = "input")
  conv1_1 = Conv2D(32, (3, 3), padding = "same", activation='relu', kernel_regularizer=regularizers.l2(l2_lambda), name = "conv1_1")(input)
  conv1_1 = bn(name = "conv1_1_bn")(conv1_1)
  conv1_2 = Conv2D(32, (3, 3), padding = "same", activation='relu', kernel_regularizer=regularizers.l2(l2_lambda), name = "conv1_2")(conv1_1)
  conv1_2 = bn(name = "conv1_2_bn")(conv1_2)
  pool1 = MaxPooling2D(name = "pool1")(conv1_2)
  drop1 = Dropout(dropout_rate)(pool1)
  
  conv2_1 = Conv2D(64, (3, 3), padding = "same", activation=LeakyReLU(alpha=0.1), kernel_regularizer=regularizers.l2(l2_lambda), name = "conv2_1")(pool1)
  conv2_1 = bn(name = "conv2_1_bn")(conv2_1)
  conv2_2 = Conv2D(64, (3, 3), padding = "same", activation=LeakyReLU(alpha=0.1), kernel_regularizer=regularizers.l2(l2_lambda), name = "conv2_2")(conv2_1)
  conv2_2 = bn(name = "conv2_2_bn")(conv2_2)
  pool2 = MaxPooling2D(name = "pool2")(conv2_2)
  drop2 = Dropout(dropout_rate)(pool2)
  
  conv3_1 = Conv2D(128, (3, 3), padding = "same", activation=LeakyReLU(alpha=0.1), kernel_regularizer=regularizers.l2(l2_lambda), name = "conv3_1")(pool2)
  conv3_1 = bn(name = "conv3_1_bn")(conv3_1)
  conv3_2 = Conv2D(128, (3, 3), padding = "same", activation=LeakyReLU(alpha=0.1), kernel_regularizer=regularizers.l2(l2_lambda), name = "conv3_2")(conv3_1)
  conv3_2 = bn(name = "conv3_2_bn")(conv3_2)
  pool3 = MaxPooling2D(name = "pool3")(conv3_2)
  drop3 = Dropout(dropout_rate)(pool3)  

  conv4_1 = Conv2D(256, (3, 3), padding = "same", activation=LeakyReLU(alpha=0.1), kernel_regularizer=regularizers.l2(l2_lambda), name = "conv4_1")(pool3)
  conv4_1 = bn(name = "conv4_1_bn")(conv4_1)
  conv4_2 = Conv2D(256, (3, 3), padding = "same", activation=LeakyReLU(alpha=0.1), kernel_regularizer=regularizers.l2(l2_lambda), name = "conv4_2")(conv4_1)
  conv4_2 = bn(name = "conv4_2_bn")(conv4_2)
  pool4 = MaxPooling2D(name = "pool4")(conv4_2)
  drop4 = Dropout(dropout_rate)(pool4)  

  conv5_1 = Conv2D(512, (3, 3), padding = "same", activation=LeakyReLU(alpha=0.1), kernel_regularizer=regularizers.l2(l2_lambda), name = "conv5_1")(pool4)
  conv5_1 = bn(name = "conv5_1_bn")(conv5_1)
  conv5_2 = Conv2D(512, (3, 3), padding = "same", activation=LeakyReLU(alpha=0.1), kernel_regularizer=regularizers.l2(l2_lambda), name = "conv5_2")(conv5_1)
  conv5_2 = bn(name = "conv5_2_bn")(conv5_2)
  
  # Decoder
  upconv6 = Conv2DTranspose(256,(2, 2), strides=(2, 2), padding='same')(conv5_2)
  upconv6 = Dropout(dropout_rate)(upconv6)
  concat6 = concatenate([conv4_2, upconv6], name = "concat6")
  conv6_1 = Conv2D(256, (3, 3), padding = "same", kernel_regularizer=regularizers.l2(l2_lambda), name = "conv6_1")(concat6)
  conv6_1 = bn(name = "conv6_1_bn")(conv6_1)
  conv6_2 = Conv2D(256, (3, 3), padding = "same", kernel_regularizer=regularizers.l2(l2_lambda), name = "conv6_2")(conv6_1)
  conv6_2 = bn(name = "conv6_2_bn")(conv6_2)
    
  upconv7 = Conv2DTranspose(128,(2, 2), strides=(2, 2), padding='same')(conv6_2)
  upconv7 = Dropout(dropout_rate)(upconv7)
  concat7 = concatenate([conv3_2, upconv7], name = "concat7")
  conv7_1 = Conv2D(128, (3, 3), padding = "same", kernel_regularizer=regularizers.l2(l2_lambda), name = "conv7_1")(concat7)
  conv7_1 = bn(name = "conv7_1_bn")(conv7_1)
  conv7_2 = Conv2D(128, (3, 3), padding = "same", kernel_regularizer=regularizers.l2(l2_lambda), name = "conv7_2")(conv7_1)
  conv7_2 = bn(name = "conv7_2_bn")(conv7_2)

  upconv8 = Conv2DTranspose(64,(2, 2), strides=(2, 2), padding='same')(conv7_2)
  upconv8 = Dropout(dropout_rate)(upconv8)
  concat8 = concatenate([conv2_2, upconv8], name = "concat8")
  conv8_1 = Conv2D(64, (3, 3), padding = "same", kernel_regularizer=regularizers.l2(l2_lambda), name = "conv8_1")(concat8)
  conv8_1 = bn(name = "conv8_1_bn")(conv8_1)
  conv8_2 = Conv2D(64, (3, 3), padding = "same", kernel_regularizer=regularizers.l2(l2_lambda), name = "conv8_2")(conv8_1)
  conv8_2 = bn(name = "conv8_2_bn")(conv8_2)

  upconv9 = Conv2DTranspose(32,(2, 2), strides=(2, 2), padding='same')(conv8_2)
  upconv9 = Dropout(dropout_rate)(upconv9)
  concat9 = concatenate([conv1_2, upconv9], name = "concat9")
  conv9_1 = Conv2D(32, (3, 3), padding = "same", kernel_regularizer=regularizers.l2(l2_lambda), name = "conv9_1")(concat9)
  conv9_1 = bn(name = "conv9_1_bn")(conv9_1)
  conv9_2 = Conv2D(32, (3, 3), padding = "same", kernel_regularizer=regularizers.l2(l2_lambda), name = "conv9_2")(conv9_1)
  conv9_2 = bn(name = "conv9_2_bn")(conv9_2)
  dropout = Dropout(dropout_rate)(conv9_2)
  
  conv10 = Conv2D(15, (1, 1), padding = "same", name = "conv10")(dropout)

 
  model = Model(input, conv10)
  
  return model
```

Functions to get optical flow and warp frames using optical flow

```python
import cv2 
import numpy
from google.colab.patches import cv2_imshow
import skimage

def get_optical_flow(prev_frame: numpy.ndarray, next_frame: numpy.ndarray) -> numpy.ndarray:
    prev_gray = cv2.cvtColor(prev_frame, cv2.COLOR_BGR2GRAY)
    next_gray = cv2.cvtColor(next_frame, cv2.COLOR_BGR2GRAY)
    flow = cv2.calcOpticalFlowFarneback(prev_gray, next_gray, None, 0.5, 3, 15, 3, 5, 1.2, 0)
    return flow

def warp_frame(prev_frame: numpy.ndarray, flow: numpy.ndarray):
    h, w = flow.shape[:2] 
    flow = -flow
    flow[:,:,0] += numpy.arange(w)
    flow[:,:,1] += numpy.arange(h)[:,numpy.newaxis]
    # print("warping function: ",flow.shape, prev_frame.shape)
    # res = cv2.remap(img, flow, None, cv2.INTER_LINEAR)
    next_frame = cv2.remap(prev_frame, flow, None, cv2.INTER_LINEAR)
    return next_frame
```

```python
import cv2
import glob
import math

X_data = []
Y_data = []
files = glob.glob ("/content/origin_video_frames/*.png")
files.sort()
# print("origin:",len(files))
for myFile in files:
    image = cv2.imread(myFile)
    # image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)#comment if frame already rgb
    X_data.append(image)
files = glob.glob ("/content/video_frames/*.png")#check path
files.sort()
# print("processed:",len(files))
# print(files)
for myFile in files:
    image = cv2.imread(myFile)
    # image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    Y_data.append(image)

imgscnt = 225                                #change value
X_data2 = np.array(X_data)[:imgscnt]                                           
X_data,Y_data = np.array(X_data)[:imgscnt]/255,np.array(Y_data)[:imgscnt]/255
# print(X_data.shape, Y_data.shape)
batch = 1
part = math.ceil(imgscnt/batch)
x = np.zeros((part, batch, 128, 128, 3))
y = np.zeros((part, batch, 128, 128, 3))

m=0
for i in range(part):
  for j in range(batch):
    x[i][j] = X_data[m]
    y[i][j] = Y_data[m]
    m+=1
    if m==imgscnt:
      break

print('x shape:', x.shape)
print('y shape:', y.shape)
```

```python
def get_vgg_features(img):
  img = np.squeeze(img,axis=0)
  invgg = cv2.resize(img, (224,224))
  invgg = preprocess_input(invgg)
  invgg = np.expand_dims(invgg/255,0)
  vgg_layers = vgg_model.layers
  output_layer=['3', '8', '15']
  output_layer = output_layer
  layer_name_mapping = {
              '3': "relu1_2",
              '8': "relu2_2",
              '15': "relu3_3",
              '22': "relu4_3"
          }
      
  feature_list = []
  if type(vgg_layers) == list:
    for layer, module in enumerate(vgg_layers):
      invgg = module(invgg)
      if str(layer) in output_layer:
        feature_list.append(invgg)
  else:
    for name, module in vgg_layers._modules.items():
      x = module(invgg)
      if name in self.output_layer:
        feature_list.append(invgg)
  return feature_list
print(len(get_vgg_features(x[10])))
```

loss_img_flow_long_term returns total loss using loss_img which is image reconstruction loss and flow loss.

perceptual loss compares vgg features for two images.

```python
def L1_loss(img1, img2):
  return K.mean(K.abs(img1 - img2)) #L2_L1 problems 

def loss_img_flow_long_term(frame_in, y_pred, flow_lambda, mask):#mask last argument

  frame_out,fpn5,fpn3,fpn1,fp1,fp3,fp5 = y_pred[:,:,:,0:3],y_pred[:,:,:,3:5],y_pred[:,:,:,5:7],y_pred[:,:,:,7:9],y_pred[:,:,:,9:11],y_pred[:,:,:,11:13],y_pred[:,:,:,13:15]
  frame_processed,fn5,fn3,fn1,f1,f3,f5 = frame_in[:,:,:,0:3],frame_in[:,:,:,3:5],frame_in[:,:,:,5:7],frame_in[:,:,:,7:9],frame_in[:,:,:,9:11],frame_in[:,:,:,11:13],frame_in[:,:,:,13:15]
  # print("loss frames and flow ",frame_out.shape, frame_processed.shape, fpn5.shape, fn5.shape)
  loss_img = L1_loss(frame_out, frame_processed)#multiply with maska both

  flowloss = L1_loss(fpn1 ,fn1)+L1_loss(fpn3 ,fn3)+L1_loss(fpn1 ,fn1)+L1_loss(fp1 ,f1)+L1_loss(fp3 ,f3)+L1_loss(fp5 ,f5)
  # flowloss = L1_loss(fpn1 ,fn1)+L1_loss(fp1 ,f1)+L1_loss(fp3,f3)+L1_loss(fpn3,fn3)+L1_loss(fp5 ,f5)
  loss = loss_img + flow_lambda*flowloss
  return loss

def perceptual_loss(img1,img2):
  img1,img2 = np.array(img1),np.array(img2)
  features1 = get_vgg_features(img1)
  features2 = get_vgg_features(img2)
  loss=0
  for i in range(3):
    loss+=L1_loss(features1[i], features2[i])
  return loss
```

check_flow_occlusion function is to calculate occlusion mask which we will use in warping error calculation.

The function is inspired from internal learning video inpainting : https://github.com/Haotianz94/IL_video_inpainting/blob/master/src/utils.py

```python
def check_flow_occlusion(flow_f, flow_b):
    """
    Compute occlusion map through forward/backward flow consistency check
    """
    def get_occlusion(flow1, flow2):
        flow1 = torch.from_numpy(flow1)
        flow2 = torch.from_numpy(flow2)
        grid_flow = grid + flow1
        grid_flow[0, :, :] = 2.0 * grid_flow[0, :, :] / max(W - 1, 1) - 1.0
        grid_flow[1, :, :] = 2.0 * grid_flow[1, :, :] / max(H - 1, 1) - 1.0
        grid_flow = grid_flow.permute(1, 2, 0)        
        flow2_inter = torch.nn.functional.grid_sample(flow2[None, ...], grid_flow[None, ...])[0]
        score = torch.exp(- torch.sum((flow1 + flow2_inter) ** 2, dim=0) / 2.)
        occlusion = (score > 0.5)
        # cv2_imshow(np.array(occlusion)*255)
        return occlusion[None, ...].float()

    H, W, C = flow_f.shape
    # Mesh grid 
    xx = torch.arange(0, W).view(1, -1).repeat(H, 1)
    yy = torch.arange(0, H).view(-1, 1).repeat(1, W)
    xx = xx.view(1, H, W)
    yy = yy.view(1, H, W)
    grid = torch.cat((xx, yy), 0).float()
    # print(grid.shape)
    f, b = np.zeros((C,H,W)), np.zeros((C,H,W))
    for i in range(C):
      f[i,:,:] = flow_f[:,:,i]
      b[i,:,:] = flow_b[:,:,i]
    occlusion_f = get_occlusion(f, b)
    occlusion_f = np.array(occlusion_f)
    # print(occlusion_f.shape)
    occ = np.zeros((H,W,1))
    occ[:,:,0] = occlusion_f[0,:,:]
    return occ
```

Below code is to calculate and store warping error.

```python
paths = glob.glob("/content/out1/*.png")
paths_sorted_out = sorted(paths, key = lambda x: int(x.split("/")[-1].split(".")[0]))
print(paths_sorted_out)
!rm -rf /content/warping_error_proc.txt
warp_T,warp1_T,warp2_T,warp1_T_mask,warp2_T_mask,warp_T_mask = 0,0,0,0,0,0
f = open("/content/warping_error_out_nosort.txt", "w+") #flow,bothdirection,vgg,epoch,adam 0.?
N = len(paths_sorted_out)
I0 = cv2.imread(paths_sorted_out[0])
for i in range(1,N):
  I2 = cv2.imread(paths_sorted_out[i])
  I1 = cv2.imread(paths_sorted_out[i-1])
  flow = get_optical_flow(I1, I2)
  flow_inv = get_optical_flow(I2, I1)
  I22 = warp_frame(I1, flow)
  mask = check_flow_occlusion(flow, flow_inv)
  # print(mask)
  I22 = I22*mask
  I2_orig = I2*mask
  diff1 = np.mean(np.abs(I2_orig - I22))#/(np.sum(mask)+0.00000000001)
  diff1_mask = diff1/(3*np.sum(mask)+0.00000000001)
  # print(diff1)
  warp_T+=diff1
  warp1_T+=diff1
  warp1_T_mask+=diff1_mask
  warp_T_mask+=diff1_mask

  flow = get_optical_flow(I0, I2)
  flow_inv = get_optical_flow(I2, I0)
  I22 = warp_frame(I0, flow)
  mask = check_flow_occlusion(flow, flow_inv)
  I22 = I22*mask
  I2_orig = I2*mask
  diff2 = np.mean(np.abs(I2_orig - I22))#/(np.sum(mask)+0.00000000001)
  diff2_mask = diff2/(3*np.sum(mask)+0.00000000001)
  # print(diff2)
  warp_T+=diff2
  warp2_T+=diff2
  warp2_T_mask+=diff2_mask
  warp_T_mask+=diff2_mask

  f.write("{0}  {1}  {2}  {3} {4} {5} {6}\n".format(i, diff1, diff2, diff1+diff2, diff1_mask, diff2_mask, diff1_mask+diff2_mask))

f.write("total: {0} {1} {2} {3} {4} {5}".format(warp1_T/N,warp2_T/N,warp_T/N,warp1_T_mask,warp2_T_mask,warp_T_mask/N))
f.close()
```
